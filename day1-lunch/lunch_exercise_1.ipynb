# Plot multiple series

```python
x2 = [2, 4, 6]
y2 = [8, 64, 216]
```

```python
fig, ax = plt.subplots()
ax.plot( x, y )
ax.plot( x2, y2, '+' )
plt.show()
```

# Annotate plot

```python
fig, ax = plt.subplots()
ax.plot( x, y, label="x^2" )
ax.plot( x2, y2, '+', label="x^3" )
ax.set_title( "Polynomials" )
ax.set_xlabel( "x" )
ax.set_ylabel( "f(x)" )
ax.legend()
plt.show()
```

# Subplot

```python
fig, ax = plt.subplots( nrows=2 )
ax[0].plot( x, y, '.' )
ax[0].set_xlim([0.0, 6.0])
ax[0].set_ylim([0, 250])
ax[1].plot( x2, y2 )
ax[1].set_xlim([0.0, 6.0])
ax[1].set_ylim([0, 250])
plt.show()
```
