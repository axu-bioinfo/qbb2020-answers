```python
import os, sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
abundance_table_df = pd.read_csv("abundance_table.tab", sep = "\t")
abundance_table_df = abundance_table_df.set_index("Genomic bins")
abundance_table_df
```

```python
bin_taxon = pd.Series(["Staphylococcus haemolyticus JCSC1435", "Enterococcus faecalis", "Leuconostoc citreum KM20", 
                       "Cutibacterium avidum 44067", "Staphylococcus epidermidis", "Staphylococcus aureus subsp. aureus",
                      "Peptoniphilaceae", "Staphylococcus lugdunensis"], name = "bin taxonomy")
abundance_bin_tax_df = abundance_table_df.set_index(bin_taxon)
abundance_bin_tax_df
```

```python
ax = sns.clustermap(abundance_bin_tax_df)
plt.title("Heatmap of abundance of samples in each bin")
plt.show()
```
