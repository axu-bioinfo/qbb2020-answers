```python
import os, sys

with open("/Users/cmdb/qbb2020-answers/day1-evening/SRR072893.sam") as samFile:
	lines = samFile.readlines()
	#count number of alignments
	count_alignment = 0
	#count number of perfect alignments
	count_perfect_alignment = 0
	#for the average of mapq values
	total_mapq = 0
	mapq_count = 0
	#count number of reads that start alignment on chromosome 2L between 10k and 20k
	chrom_2L_count = 0
	for line in lines:
		#count number of alignments
		line = line.strip()
		split_line = line.split("\t")
		#The alignment result is shown in FLAG format in split_line[1]
		if "4" in split_line[1]:
			count_alignment = count_alignment + 0
		else:
			count_alignment = count_alignment + 1

		#count the number of alignments that match perfectly to the genome
		if "NM:i:0" in line:
			count_perfect_alignment = count_perfect_alignment + 1

		#get a list of the first 10 chromosomes
		first_ten = 0
		first_ten_chromosome_list = []
		while first_ten < 10:
			first_ten_chromosome_list.append(split_line[2])
			first_ten = first_ten + 1

		#calculate the average mapq score across all reads (mapq is found in split_line[4])
		total_mapq = int(split_line[4]) + total_mapq
		mapq_count = mapq_count + 1

		#count number of reads that start their alignment on chromosome 2L between base 10000 and 20000
		if "2L" == split_line[2]:
			if int(split_line[3]) >= 10000 and int(split_line[3]) <= 20000:
				chrom_2L_count = chrom_2L_count + 1


#first question
print("Count alignment is " + str(count_alignment) + "\n")

#second question
print("Number of perfect alignments is " + str(count_perfect_alignment) + "\n")

#third question
print("First ten reads aligned to chromosome:")
for count in range(len(first_ten_chromosome_list)):
	number_count = str(count + 1)
	print(number_count + ".\t" + first_ten_chromosome_list[count])

#fourth question
mapq_average = total_mapq/mapq_count
print("Average MAPQ scores across all reads = " + str(mapq_average) + "\n")

#fifth question
print("Total number of reads that start their alignment on chromosome 2L between 10k and 20k base inclusive: " + str(chrom_2L_count) + "\n")

#------------------------------------------------------------------------------------------------------------------------------------------------
#Final output:
#Count alignment is 2000000

#Number of perfect alignments is 1702765

#First ten reads aligned to chromosome:
#1.	2L
#2.	2L
#3.	2L
#4.	2L
#5.	2L
#6.	2L
#7.	2L
#8.	2L
#9.	2L
#10.	2L
#Average MAPQ scores across all reads = 59.6407825

#Total number of reads that start their alignment on chromosome 2L between 10k and 20k base inclusive: 7550
#------------------------------------------------------------------------------------------------------------------------------------------------
```
